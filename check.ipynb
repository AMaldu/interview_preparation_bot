```python
from nltk.tokenize import sent_tokenize

text = "Aquí tienes una oración. Aquí hay otra."
sentences = sent_tokenize(text)
print(sentences)  # Deberías ver: ['Aquí tienes una oración.', 'Aquí hay otra.']
```

```python
import spacy

nlp = spacy.load("es_core_news_sm")  # Carga el modelo de español

text = "Aquí tienes una oración. Aquí hay otra."
doc = nlp(text)
sentences = [sent.text for sent in doc.sents]  # Tokeniza en oraciones
print(sentences)
```
